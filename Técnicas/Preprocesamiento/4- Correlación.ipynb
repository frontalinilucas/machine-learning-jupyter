# Correlación
La correlación cruzada (o a veces denominada "covarianza cruzada") es una medida de la similitud entre dos señales, frecuentemente usada para encontrar características relevantes en una señal desconocida por medio de la comparación con otra que sí se conoce.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Defino la señal desconocida:
x=np.random.randn(20000)

#Defino la señal conocida:
y=np.random.randn(2000)

x[10000:12000]+=0.2*y
corr=np.correlate(x,y)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x)
axarr[0].set_title('Sharing X axis')
axarr[1].plot(y)
axarr[2].plot(corr)
plt.show()

print("El máximo de la correlación está en:")
print(corr.argmax())
```

```python
alfa=0.1

#Defino la señal desconocida:
x=np.random.randn(20000)

#Defino la señal conocida:
y=np.sin(np.linspace(0,16*np.pi,4000))

x[10000:14000]+=alfa*y
corr=np.correlate(x,y)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x)
axarr[0].set_title('Sharing X axis')
axarr[1].plot(y)
axarr[2].plot(corr)
plt.show()

print("El máximo de la correlación está en:")
print(corr.argmax())
```
